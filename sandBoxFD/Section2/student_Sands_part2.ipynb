## Baseline Modelling

```python
df_train.columns
```

```python
base = ['g1', 'g2', 'dalc', 'walc', 'goout', 'freetime' , 'famrel', 'studytime', 'traveltime','failures','medu', 'fedu', 'health', 'absences', 'age']
```

```python
df[base]
```

```python
def train_linear_regression(X,y,r):
    ones = np.ones(X.shape[0])
    X = np.column_stack([ones, X])
    XTX = X.T.dot(X)
    XTX = XTX + r* np.eye(XTX.shape[0])
    XTX_inv = np.linalg.inv(XTX)
    w_full = XTX_inv.dot(X.T).dot(y)
    return w_full[0], w_full[1:]
```

```python
def prepare_X(df):
    df_num = df[base]
    X = df_num.values
    return X
```

```python
X_train = prepare_X(df_train)
```

```python
w0, w = train_linear_regression(X_train, y_train, r = 10)
```

```python
y_pred = w0 + X_train.dot(w)
```

sns.histplot(y_pred, color='red')
sns.histplot(y_train, color = 'blue')

```python
def rmse(y, y_pred):
    se = (y - y_pred) ** 2
    mse = se.mean()
    return np.sqrt(mse)
```

```python
rmse(y_train, y_pred)
```

## Validatiton for only 2 

```python
X_val = prepare_X(df_val)
y_pred = w0 + X_val.dot(w)

rmse(y_val, y_pred)
```

```python
sns.histplot(y_pred, color='red')
sns.histplot(y_val, color = 'blue')
```

```python
for r in [0.0, 0.1, 0.01, 0.001,0.00001, 10, 1000, 100000, 0.0001, 0.00001]:
    X_train = prepare_X(df_train)
    w0, w = train_linear_regression(X_train, y_train, r= r)

    X_val = prepare_X(df_val)
    y_pred = w0 + X_val.dot(w)

    score = rmse(y_val,y_pred)
    print (r, w0, score)
```
